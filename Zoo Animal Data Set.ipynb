# Zoo Animal Data Set

source: https://www.kaggle.com/uciml/zoo-animal-classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('archive/zoo.csv')
df2 = pd.read_csv('archive/class.csv')
```

```python
df2.head()
```

```python
df.describe()
```

```python
df2.describe()
```

```python
df = pd.merge(df, df2, how = 'left', left_on = 'class_type', right_on = 'Class_Number')
```

```python
df.head()
```

## 1. How many animal types are there?

```python
sns.catplot('Class_Type', data = df, kind = 'count', aspect = 2)
```

There are 7 animal types: mammal, fish, bird, invertebrate, bug, amphibian, and reptile.

## 2. Are there any correlations between each features?

```python
corr = df.iloc[:,1:-1].corr()
colormap = sns.diverging_palette(220, 10, as_cmap = True)
plt.figure(figsize=(14, 14))
sns.heatmap(corr, cbar = True,  square = True, annot = True, fmt = '.2f', annot_kws = {'size': 12}, cmap = colormap, linewidths = 0.1, linecolor = 'white')
plt.title('Correlation of ZOO Features', y = 1.05, size = 15)
```

Animals that lay egg probably will not produce milk and unlikely to have hair.

## 3. How many mammals lay egg?

```python
mammals = df.loc[df['class_type'] == 1]
mammals_eggs = mammals.loc[mammals['eggs'] == 1]
print('There is/are ' + str(len(mammals_eggs)) + ' mammal(s) that lay eggs.')
print(mammals_eggs['animal_name'])
```

## 4. How many mammals are airborne?

```python
mammals_air = mammals.loc[mammals['airborne'] == 1]
print('There is/are ' + str(len(mammals_air)) + ' mammal(s) that is/are airborne.')
print(mammals_air['animal_name'])
```

## 5. How many mammals are aquatic?

```python
mammals_aqua = mammals.loc[mammals['aquatic'] == 1]
print('There is/are ' + str(len(mammals_aqua)) + ' mammal(s) that is/are aquatic.')
print(mammals_aqua['animal_name'])
```

## Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
x = df.iloc[:,1:17]
y = df.iloc[:,17]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1, stratify = y)
```

```python
lr = LogisticRegression(multi_class = 'multinomial', solver = 'newton-cg')
score_lr = cross_val_score(lr, x, y, cv = 5)
print('The accuracy of the logistic regression model is ' + str(score_lr.mean()))
```
